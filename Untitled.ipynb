```python
# %% [markdown]
# # Kenya Child Health & Nutrition Dashboard Engine
# This notebook cleans, processes, and structures child health and nutrition data 
# from the World Bank country dataset.

# %%
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

# 1. Load the raw data (skipping the metadata headers)
df_raw = pd.read_csv("API_KEN_DS2_en_csv_v2_5938.csv", skiprows=4)

# 2. Define our target health and nutrition indicators
target_indicators = {
    "Mortality rate, under-5 (per 1,000 live births)": "Child Mortality Rate",
    "Prevalence of stunting, height for age (% of children under 5)": "Child Stunting Rate",
    "Prevalence of underweight, weight for age (% of children under 5)": "Child Underweight Rate"
}

# 3. Filter for our indicators
df_health = df_raw[df_raw["Indicator Name"].isin(target_indicators.keys())].copy()

# 4. Map friendly, clean indicator names
df_health["Friendly Name"] = df_health["Indicator Name"].map(target_indicators)

# 5. Melt the wide year columns into long-form dataframe
year_cols = [col for col in df_health.columns if col.isdigit()]
df_long = df_health.melt(
    id_vars=["Friendly Name"], 
    value_vars=year_cols, 
    var_name="Year", 
    value_name="Value"
)

# Convert types for strict accuracy
df_long["Year"] = df_long["Year"].astype(int)
df_long["Value"] = pd.to_numeric(df_long["Value"], errors="coerce")

# Drop null values so Plotly draws continuous, unbroken timelines
df_clean = df_long.dropna(subset=["Value"])

print("Data cleaning complete! Sample rows:")
display(df_clean.head())
```

```python
# %%
# Pivot the dataset so we can compare variables on the same year row easily
df_pivot = df_clean.pivot(index="Year", columns="Friendly Name", values="Value").reset_index()

# Create a high-fidelity double-axis visual
fig = go.Figure()

# Add the Child Mortality Line (Continuous Timeline)
fig.add_trace(
    go.Scatter(
        x=df_pivot["Year"],
        y=df_pivot["Child Mortality Rate"],
        name="Child Mortality (per 1K)",
        line=dict(color="#DC2626", width=3), # Dark Crimson Red
        yaxis="y1"
    )
)

# Add Child Stunting Columns (Survey-specific years)
fig.add_trace(
    go.Bar(
        x=df_pivot["Year"],
        y=df_pivot["Child Stunting Rate"],
        name="Child Stunting Rate (%)",
        marker_color="#16A34A", # Health Green
        opacity=0.7,
        yaxis="y2"
    )
)

# Set up the double Y-axes layout for perfect scale matching
fig.update_layout(
    title="<b>The Nutrition-Survival Loop (Kenya WDI)</b>",
    xaxis=dict(title="Year", gridcolor="#E2E8F0"),
    yaxis=dict(
        title="Under-5 Mortality Rate (per 1,000)",
        tickfont=dict(color="#DC2626"),
        gridcolor="#E2E8F0"
    ),
    yaxis2=dict(
        title="Stunting Prevalence (% of children under 5)",
        tickfont=dict(color="#16A34A"),
        overlaying="y",
        side="right"
    ),
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    plot_bgcolor="rgba(0,0,0,0)", # Transparent background
    paper_bgcolor="rgba(0,0,0,0)"
)

fig.show()
```
